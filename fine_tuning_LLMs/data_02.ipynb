```python
import json
import pandas as pd
```

```python
# Load the JSON data from the uploaded file
with open('.//KB.json', 'r') as file:
    data = json.load(file)

# Create a list to store dictionaries of pattern-response pairs
pattern_response_pairs = []

# Iterate over each intent to create pairs of pattern and response
for intent in data['intents']:
    tag = intent.get('tag', '')
    patterns = intent.get('patterns', [])
    responses = intent.get('responses', [])
    
    for pattern in patterns:
        for response in responses:
            pattern_response_pairs.append({'tag': tag, 'user': pattern, 'assistant': response})

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(pattern_response_pairs)
df
```

```python
df.to_csv('./data/data_02.csv', index=False)
```
