```python
import pandas as pd
import re
```

```python
file_path = './train_02.csv'
df = pd.read_csv(file_path)
df
```

```python
def split_dialogue(text):
    human_text = ''
    assistant_text = ''
    
    human_match = re.search(r'<HUMAN>:(.*?)<ASSISTANT>:', text, re.DOTALL)
    assistant_match = re.search(r'<ASSISTANT>:(.*)', text, re.DOTALL)
    
    if human_match:
        human_text = human_match.group(1).strip()
    if assistant_match:
        assistant_text = assistant_match.group(1).strip()
    
    return pd.Series([human_text, assistant_text], index=['user', 'assistant'])

df = df['text'].apply(split_dialogue)
df
```

```python
df.to_csv('./data/data_04.csv', index=False)
```
